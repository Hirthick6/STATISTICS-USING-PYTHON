```python
import pandas as pd
import numpy as np
df = pd.read_excel('restaurant.xlsx')
df
```

# Cross Tabulation

```python
pd.crosstab(df.Quality_rating,df.Meal_Price)
```

```python
pd.crosstab(df.Quality_rating,df.Meal_Price,margins=True)
```

# Create a cross tab with percentages

```python
crosstb2 = pd.crosstab(index = df['Quality_rating'],  \
                      columns = df['Meal_Price'] , \
                      normalize=True) *100
crosstb2
```

# Adding a total column and row to the crosstab

```python
crosstb3 = pd.crosstab(index = df['Quality_rating'],  \
                      columns = df['Meal_Price'] , \
                      normalize=True, margins = True, \
                      margins_name= "Total") *100
crosstb3
```

# Percentage Frequency

```python
crosstb3.astype(str).apply(lambda x:x + '%')
```
