```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
raw_data=pd.read_csv('Salary_Data.csv')
a=raw_data['x']
b=raw_data['y']
ma=np.mean(a)
nb=np.mean(b)
print("Mean of a: ",ma)
print("Mean of b: ",nb)
n=len(a)
nu=0
de=0
for i in range(n):
    nu+=((a[i]-ma)*(b[i]-nb))
    de+=(a[i]-ma)**2
m=nu/de
c=nb-(m*ma)
print("Value of m: ",m,"\nValue of c: ",c)
```

```python
maxx=np.max(a)
minx=np.min(a)
x=np.linspace(minx,maxx,10)
y=c+m*x
```

```python
plt.plot(x,y,color='black',label='Regression Line')
plt.scatter(a,b,color='red',label='Scatter Plot')
plt.xlabel('x-independent variable')
plt.ylabel('y-dependant variable')
plt.legend()
plt.show()
```

```python
sr=0
st=0
for i in range(n):
    yp=m*a[i]+c
    sr+=(yp-nb)**2
    st+=(b[i]-nb)**2
r2=(sr/st)
print("Value of r2: ",r2)
```
